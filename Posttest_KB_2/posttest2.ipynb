#### Ketentuan 3 
Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.

```python
from prettytable import PrettyTable #mengimport prettytable, untuk membuat tampilan table lebih rapi

jr = film.size
ja = len(film.columns)
nm = film.columns
jangka = len(film.select_dtypes('int').columns)

info = {
    "Jumlah Record" : jr,
    "Jumlah Attribute" : ja,
    "Nama Attribute" : nm,
    "Jumlah Attribute bertipe Angka" : jangka
}

table = PrettyTable()
for i in info.keys():
    table.add_column(i, [])
table.add_row([info.get(i, "") for i in info.keys()])
print(table)
```
