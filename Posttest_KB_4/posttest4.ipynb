## Data Cleaning

Lakukan Data cleaning pada data dengan nilai null (jika tidak ada nilai null pada dataset, maka buatlah menjadi ada). Ganti nilai null sesuai ketentuan. (bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus). Nilai plus jika strategy yang digunakan menggunakan 3 nilai

```python
fc2.info()
```

### pada nilai null

```python
print ('Jumlah record yang memiliki nilai null : ') # melihat nilai null
print (fc2.isnull().sum())
```

```python
fc2['metascore'].fillna(fc2['metascore'].median(), inplace = True)
fc2.isna().sum()
```

```python
fc2['gross'].fillna(fc2['gross'].mean(), inplace = True)
fc2.isna().sum()
```

```python
imputer = SimpleImputer(strategy='most_frequent')
fc2['certificate'] = imputer.fit_transform(fc2[['certificate']])
fc2.isna().sum()
```
